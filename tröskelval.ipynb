Denna notebook används för val av tröskelvärden inom Extreme Value Theory (EVT)
för svenska aktieavkastningar med hjälp av Peaks-over-threshold-metoden.

## Hämta paket

```python
import yfinance as yf           # Ladda ner historisk data från Yahoo Finance
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import pyextremes as extremes   # Huvudpaket för extreme value analysis (EVA)

from pandas_datareader import data as web
from scipy import stats         # Statistiska distributioner och hypotes tester

from pyextremes import get_extremes             # Extrahera extrema observationer över en tröskel
from pyextremes.plotting import plot_extremes   # Visualisera extrema händelser
from pyextremes import plot_mean_residual_life  # MRL plot för tröskelval
from pyextremes import plot_parameter_stability # Parameterstabilitet diagnostik för tröskelval
from pyextremes import EVA                      # Huvudklass för modellering av extremvärden
```

## Ladda ner data

```python
# ladda ner data från företag på Stockholm Stock Exchange (.ST)
start = "2000-01-01"
end = "2026-03-31"

# daglig data från svenska börsnoterade bolag, inklusive OMXS:
data = {
    "Kinnevik B": yf.download("KINV-B.ST", start=start, end = end,interval="1d", auto_adjust=True, progress=False),
    "Investor B": yf.download("INVE-B.ST", start=start, end = end,interval="1d", auto_adjust=True, progress=False),
    "SEB A": yf.download("SEB-A.ST", start=start, end = end,interval="1d", auto_adjust=True, progress=False),
    "Nordea Bank": yf.download("NDA-SE.ST", start=start, end = end, interval="1d", auto_adjust=True, progress=False),
    "Castellum": yf.download("CAST.ST", start=start, end = end,interval="1d", auto_adjust=True, progress=False),
    "Skanska B": yf.download("SKA-B.ST", start=start, end = end,interval="1d", auto_adjust=True, progress=False),
    "ABB": yf.download("ABB.ST", start=start, end = end,interval="1d", auto_adjust=True, progress=False),
    "Atlas Copco A": yf.download("ATCO-A.ST", start=start,end = end, interval="1d", auto_adjust=True, progress=False),
    "Boliden": yf.download("BOL.ST", start=start, end = end,interval="1d", auto_adjust=True, progress=False),
    "SCA B": yf.download("SCA-B.ST", start=start, end = end, interval="1d", auto_adjust=True, progress=False)
}
# auto_adjust används för att justera aktiepriser för utdelningar och splittar
# OMXS är ett prisindex, så någon motsvarande justering krävs inte


portföljer = {
    "Investment" : ["Industrivärden C", "Investor B",  "Kinnevik B"], # dessa representerar i princip indexen
    "Bank" : ["Handelsbanken A", "Nordea Bank", "SEB A", "Swedbank A"], # följer i princip samma graf så kanske ointressant
    "Fastigheter" : ["Skanska B"],
    "Industri": ["Assa Abloy B", "Atlas Copco A", "Sandvik", "Volvo B"]
}
```

```python
# välj nu enbart ut de datum som förekommer i alla dataframes i vår data-dictionary

common_idx = sorted(set.intersection(*(set(df.index) for df in data.values())))
# Koden hittar alla index som förekommer i alla DataFrames
# i data och returnerar dem som en sorterad lista.


# definiera en ny data-dictionary med samma keys men med enbart matchande index
data_clean = {}
for k in data:
    data_clean[k] = data[k].loc[common_idx].copy()
for df in data_clean.values():
    df['lgr'] = np.log(df['Close']).diff() # Skapa kolumn för log returns
```

## Initiera tabell för att senare spara tröskelvärden

```python
cols = [
    "Threshold",
    "Quantile",             # (valfri) emp. kvantil för u (notera att denna tar med alla överskridande i ett kluster)
    "Number of Exceedences",
    "Type",
    "Rate",                 # (valfri) exceedance rate = n_exc / n_obs
    "Shape xi",             
    "Scale beta",
    "Modified scale",
    "QQ R2",
    "PP R2",
    "Run Length r",
]
thresh_summary = pd.DataFrame(columns=cols)
thresh_summary.index.name = "Stock"
```

```python
df = pd.DataFrame()

# Extrahera serien med logaritmiska avkastningar ('lgr') från varje rensad dataserie
for key in data_clean.keys():
    df[key] = data_clean[key]['lgr']

# Lista över de valda aktierna
stocks = [
    "Kinnevik B",
    "Investor B",
    "SEB A",
    "Nordea Bank",
    "Castellum",
    "Skanska B",
    "ABB",
    "Atlas Copco A",
    "Boliden",
    "SCA B"
]
# Skapa en likaviktad portfölj genom att beräkna medelvärdet av aktiernas avkastningar
df["Portfölj"] = df[stocks].mean(axis=1)
# Ordna kolumnerna så att portföljen placeras sist
df = df[stocks + ["Portfölj"]]

data = df[key].dropna()
series = -df[key].dropna() # Förluster blir positiva
```

### $r$ i $r$-runs-metoden och intervallet av tröskelvärden i MRL ska vara samma för alla aktier

```python
r='15D'                           
thresh = np.linspace(0.02, 0.10)
```

## "Key" måste ändras varje gång vi kollar en ny aktie

```python
key = 'Portfölj' # Välj vilken tillgång som ska analyseras

plot_mean_residual_life(series, thresh)
plt.title(key)
plt.show()
```

```python
plot_parameter_stability(series, thresh)
plt.show()
```

### Spara ett tröskelvärde som anses passa bra från ovanstående grafer i "threshold"

```python
model = EVA(series)

threshold = 0.032

model.get_extremes(method="POT", r=r, threshold = threshold, extremes_type = 'high')
model.plot_extremes()
plt.show()
```

```python
model.fit_model()

summary = model.get_summary(return_period_size = "365D",
    return_period=np.linspace(1,325),
    alpha=0.95,
    n_samples=1000,
)

model.plot_diagnostic(alpha = 0.95)
plt.show()
```

```python
len(model.extremes) # Vi vill åtminstone ha 50 datapunkter som överskrider
```

### Om $u$ verkar passa bra sparas både Q-Q och P-P plottens $R^2$ värden

```python
QQ_R2 = 0.995 
PP_R2 = 0.999
```

## Efter flera trösklar har testats, sparas värdena genom att köra cellen nedan

```python
u = threshold
dist = model.distribution
params = dist.mle_parameters
quantile = (series <= u).mean()

if dist.name == "expon":
    xi = 0.0
    beta = float(params["scale"])
    type = "Exponential"
elif dist.name == "genpareto":
    xi = float(params["c"])       # shape
    beta = float(params["scale"]) # scale
    type = "GPD"
else:
    raise ValueError(f"Okänd POT-distribution: {dist.name}")

thresh_summary.loc[key] = {
    "Threshold": u,
    "Quantile": quantile,
    "Number of Exceedences": len(model.extremes),
    "Type": type,
    "Rate": len(model.extremes) / len(series),
    "Shape xi": xi,
    "Scale beta": beta,
    "Modified scale": beta + xi*u,
    "QQ R2": QQ_R2,
    "PP R2": PP_R2,
    "Run Length r": r 
}
```
